```python
notes = [
    "Patient reports taking metformin daily but seems uncertain about dosage. Concerned about side effects.",
    "Patient states medication is working well. No complaints. Follow up in 3 months.",
    "Patient admits missing doses occasionally. Not sure if insulin is helping. Anxious about blood sugar levels.",
    "Prescription unchanged. Patient appears confident and well informed about treatment plan.",
    "Patient frustrated with lack of improvement. Mentioned stopping medication last week without informing doctor."
]

for i, note in enumerate(notes):
    print(f"Note {i+1}: {note}")
    print()
```

```python
concern_words = ["uncertain", "anxious", "concerned", "frustrated", 
                 "missing", "stopping", "unsure", "worried"]

print("PATIENT CONCERN DETECTION:")
print("-" * 40)

for i, note in enumerate(notes):
    found = [word for word in concern_words if word.lower() in note.lower()]
    risk = "🔴 HIGH RISK" if len(found) >= 2 else "🟡 MONITOR" if len(found) == 1 else "🟢 LOW RISK"
    print(f"Note {i+1}: {risk}")
    print(f"  Concern signals found: {found}")
    print()
```
